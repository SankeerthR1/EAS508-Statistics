```python
import subprocess

NOTEBOOKS = [
    "scripts/download_futures.ipynb",
    "scripts/download_macro_fred.ipynb",
    "scripts/download_climate_disaster.ipynb",
    "scripts/build_monthly_panel.ipynb",
]

if __name__ == "__main__":
    for notebook in NOTEBOOKS:
        cmd = [
            "jupyter",
            "nbconvert",
            "--to",
            "notebook",
            "--execute",
            notebook,
            "--inplace",
        ]
        print(f"Running: {' '.join(cmd)}")
        subprocess.run(cmd, check=True)

```
